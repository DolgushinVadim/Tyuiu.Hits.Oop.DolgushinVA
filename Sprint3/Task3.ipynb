Описание задачи:

Создать базовый класс Book в C#, который будет представлять информацию о
книгах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Book:

• Атрибуты: Название (Title), Автор (Author), Год издания (YearOfPublication).

• Методы:

o GetInfo(): метод для получения информации о книге в виде строки.

o Read(): метод для вывода сообщения о чтении книги.

o Borrow(): метод для вывода сообщения о выдаче книги на чтение.

<h2 style="color:DodgerBlue">Требования к производным классам:</h2>

----

1. Учебник (Textbook): Должен содержать дополнительные атрибуты, такие как
Предмет (Subject). Метод Read() должен быть переопределен для
добавления информации о предмете при чтении учебника.

2. Художественная литература (Fiction): Должен содержать дополнительные
атрибуты, такие как Жанр (Genre). Метод Borrow() должен быть
переопределен для добавления информации о жанре при выдаче книги на
чтение.

3. Научная литература (ScientificLiterature) (если требуется третий класс):
Должен содержать дополнительные атрибуты, такие как Область науки
(FieldOfScience). Метод GetInfo() должен быть переопределен для включения
информации об области науки в описании книги.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

<h2 style="color:DodgerBlue">Реализация:</h2>

----

```
using System;
using System.Collections.Generic;

public interface IBook
{
    string Title { get; }
    string Author { get; }
    int YearOfPublication { get; }
    string Publisher { get; }
    int Pages { get; }
    bool IsAvailable { get; set; }
    bool IsReserved { get; set; }
    bool IsBorrowed { get; set; }
    void GetInfo();
    void Read();
    void Borrow();
    void Borrow(string borrowerName);
    void Return();
    void Reserve(int days);
}

public interface ITextbook
{
    void Study();
}

public interface IFiction
{
    void Analyze();
}

public interface IScientificLiterature
{
    void ReviewLiterature();
}

public interface IEncyclopedia
{
    void FindInformation();
}

public class Book : IBook
{
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    public string Publisher { get; set; }
    public int Pages { get; set; }
    public bool IsAvailable { get; set; }
    public bool IsReserved { get; set; }
    public bool IsBorrowed { get; set; }

    public Book(string title, string author, int yearOfPublication, string publisher, int pages)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
        Publisher = publisher;
        Pages = pages;
        IsAvailable = true;
        IsReserved = false;
        IsBorrowed = false;
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Название: {Title}, Автор: {Author}, Год издания: {YearOfPublication}, Издатель: {Publisher}, Страниц: {Pages}");
    }

    public virtual void Read()
    {
        if (IsBorrowed)
            Console.WriteLine($"Вы читаете книгу '{Title}'");
        else
            Console.WriteLine($"Книга '{Title}' не была взята. Сначала возьмите её, чтобы читать.");
    }

    public virtual void Borrow()
    {
        if (IsAvailable)
        {
            Console.WriteLine($"Вам выдали книгу '{Title}'");
            IsAvailable = false;
            IsBorrowed = true;
        }
        else
        {
            Console.WriteLine($"Книга '{Title}' недоступна.");
        }
    }

    public virtual void Borrow(string borrowerName)
    {
        if (IsAvailable)
        {
            Console.WriteLine($"{borrowerName} взял(а) книгу '{Title}'");
            IsAvailable = false;
            IsBorrowed = true;
            IsReserved = false;
        }
        else
        {
            Console.WriteLine($"Книга '{Title}' недоступна.");
        }
    }

    public virtual void Return()
    {
        if (IsBorrowed)
        {
            Console.WriteLine($"Вы вернули книгу '{Title}'");
            IsAvailable = true;
            IsBorrowed = false;
            IsReserved = false;
        }
        else
        {
            Console.WriteLine($"Книга '{Title}' не была взята, поэтому её нельзя вернуть.");
        }
    }

    public virtual void Reserve(int days)
    {
        if (!IsAvailable && !IsReserved)
        {
            Console.WriteLine($"Книга '{Title}' зарезервирована на {days} дней.");
            IsReserved = true;
        }
        else if (IsAvailable)
        {
            Console.WriteLine($"Книга '{Title}' доступна и не может быть зарезервирована.");
        }
        else
        {
            Console.WriteLine($"Книга '{Title}' уже зарезервирована.");
        }
    }
}

public class Textbook : Book, ITextbook
{
    public string Subject { get; set; }

    public Textbook(string title, string author, int yearOfPublication, string publisher, int pages, string subject)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        Subject = subject;
    }

    public override void Read()
    {
        if (IsBorrowed)
            Console.WriteLine($"Вы читаете учебник '{Title}' по предмету '{Subject}'");
        else
            Console.WriteLine($"Учебник '{Title}' не был взят. Сначала возьмите его.");
    }

    public void Study()
    {
        if (IsBorrowed)
            Console.WriteLine($"Вы изучаете учебник '{Title}' по предмету '{Subject}'");
        else
            Console.WriteLine($"Учебник '{Title}' не был взят. Сначала возьмите его.");
    }
}

public class Fiction : Book, IFiction
{
    public string Genre { get; set; }

    public Fiction(string title, string author, int yearOfPublication, string publisher, int pages, string genre)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        Genre = genre;
    }

    public override void Borrow()
    {
        if (IsAvailable)
        {
            Console.WriteLine($"Вам выдали художественную книгу '{Title}', жанр: '{Genre}'");
            IsAvailable = false;
            IsBorrowed = true;
        }
        else
        {
            Console.WriteLine($"Книга '{Title}' недоступна.");
        }
    }

    public void Analyze()
    {
        if (IsBorrowed)
            Console.WriteLine($"Вы анализируете прочитанную главу книги '{Title}' жанра '{Genre}'");
        else
            Console.WriteLine($"Книга '{Title}' не была взята. Сначала возьмите её, чтобы анализировать.");
    }
}

public class ScientificLiterature : Book, IScientificLiterature
{
    public string FieldOfScience { get; set; }

    public ScientificLiterature(string title, string author, int yearOfPublication, string publisher, int pages, string fieldOfScience)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        FieldOfScience = fieldOfScience;
    }

    public override void GetInfo() 
    {
        Console.WriteLine($"Название: {Title}, Автор: {Author}, Год издания: {YearOfPublication}, Издатель: {Publisher}, Страниц: {Pages}, Область науки: {FieldOfScience}");
    }


    public void ReviewLiterature()
    {
        if (IsBorrowed)
            Console.WriteLine($"Вы проводите обзор литературы по теме '{FieldOfScience}' на основе книги '{Title}'");
        else
            Console.WriteLine($"Книга '{Title}' не была взята. Сначала возьмите её, чтобы проводить обзор литературы.");
    }
}

public class Encyclopedia : Book, IScientificLiterature, IEncyclopedia
{
    public string FieldOfScience { get; set; }

    public Encyclopedia(string title, string author, int yearOfPublication, string publisher, int pages, string fieldOfScience)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        FieldOfScience = fieldOfScience;
    }

    public override void GetInfo() 
    {
        Console.WriteLine($"Название: {Title}, Автор: {Author}, Год издания: {YearOfPublication}, Издатель: {Publisher}, Страниц: {Pages}, Область науки: {FieldOfScience}");
    }
    public void ReviewLiterature()
    {
        if (IsBorrowed)
        {
            Console.WriteLine($"Обзор литературы по теме '{FieldOfScience}' из энциклопедии '{Title}'");
        }
        else
        {
            Console.WriteLine($"Энциклопедия '{Title}' не была взята. Сначала возьмите её.");
        }
    }

    public void FindInformation()
    {
        if (IsBorrowed)
        {
            Console.WriteLine($"Вы ищете информацию в энциклопедии '{Title}'");
        }
        else
        {
            Console.WriteLine($"Энциклопедия '{Title}' не была взята. Сначала возьмите её.");
        }
    }
}

public class Library<T> where T : IBook
{
    private List<T> books = new List<T>();

    public void AddBook(T book)
    {
        books.Add(book);
        Console.WriteLine($"Книга '{book.Title}' добавлена в библиотеку.");
    }

    public void RemoveBook(T book)
    {
        books.Remove(book);
        Console.WriteLine($"Книга '{book.Title}' удалена из библиотеки.");
    }

    public void ShowAllBooks()
    {
        Console.WriteLine("Список всех книг в библиотеке:");
        foreach (var book in books)
        {
            book.GetInfo();
        }
    }
}

Textbook mathTextbook = new Textbook("Алгебра и начала математического анализа 11 класс", "Колягин Ю. М.", 2014, "Просвещение", 320, "Математика");
Fiction novel = new Fiction("Война и мир", "Толстой Л.Н.", 1869, "Эксмо", 1200, "Исторический роман");
ScientificLiterature physicsBook = new ScientificLiterature("Классическая механика", "Ньютон И.С.", 1687, "Наука", 450, "Физика");
Encyclopedia scienceEncyclopedia = new Encyclopedia("Большая советская энциклопедия", "Коллектив авторов под редакцией Отто Ю. Шмидта", 1978, "Советская энциклопедия", 3000, "Общественные и естественные науки, охватывающие широкий спектр знаний.");

Library<Book> library = new Library<Book>();

library.AddBook(mathTextbook);
library.AddBook(novel);
library.AddBook(physicsBook);
library.AddBook(scienceEncyclopedia);
library.ShowAllBooks();

physicsBook.GetInfo();
physicsBook.Borrow();
physicsBook.Read();
physicsBook.ReviewLiterature();
physicsBook.Return();

scienceEncyclopedia.GetInfo();
scienceEncyclopedia.Borrow("Буранов Алексей Михайлович");
scienceEncyclopedia.Reserve(7);
```
